```python
# visual representation of heatmap on all the numerical variables
corr_matrix = data.corr()
plt.figure(figsize=(20,10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation heatmap of numerical variables")
plt.show()
```

### Categorical Variables  visualization

```python
sns.histplot(data['OverTime'], kde=False)
plt.title("Distribution of OverTime")
plt.show()
```

```python
sns.histplot(data['MaritalStatus'], kde=False)
plt.title("Distribution of Marital Status")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=data, y = 'JobRole')
plt.title('Job Role')
plt.show()
```
